```python
import pandas as pd
from textblob import TextBlob

def analyze_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return "positive"
    elif analysis.sentiment.polarity < 0:
        return "negative"
    else:
        return "neutral"

def main():
    # Replace 'your_file.csv' with the actual CSV file path
    csv_file = 'test.csv'
    column_name = 'text'  # Replace with the actual column name

    try:
        df = pd.read_csv(csv_file,encoding='iso-8859-1')
        if column_name not in df.columns:
            print(f"Column '{column_name}' not found in the CSV file.")
            return
        df[column_name] = df[column_name].astype(str)
        df['generated_sentiment'] = df[column_name].apply(analyze_sentiment)
        
        # Save the updated DataFrame with sentiment analysis results
        df.to_csv('output_with_sentiment.csv', index=False)
        
        print("Sentiment analysis completed and saved to 'output_with_sentiment.csv'.")

    except Exception as e:
        print("An error occurred:", str(e))

if __name__ == "__main__":
    main()
```

```python

import pandas as pd

def main():
    # Load the CSV data into a pandas DataFrame
    data = pd.read_csv("output_with_sentiment.csv")

    # Count the total number of rows
    total_rows = len(data)

    # Count the number of rows where sentiment matches generated_sentiment
    match_count = (data["sentiment"] == data["generated_sentiment"]).sum()

    # Calculate the percentage of matches
    match_percentage = (match_count / total_rows) * 100

    print("Total rows:", total_rows)
    print("Matching rows:", match_count)
    print("Match percentage:", match_percentage)

if __name__ == "__main__":
    main()
```
